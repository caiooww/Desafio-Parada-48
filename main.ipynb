```python
#Uma nova planilha mas com diferentes focos

diasUteis = planilha.iloc[0:5]
fimDeSemana = planilha.iloc[5:6]


display(diasUteis)
display(fimDeSemana)
```

Calculando o total de Alunos faltantes

```python
TotalFaltas = fimDeSemana["Alunos com falta"].sum() + diasUteis["Alunos com falta"].sum()

print("Total de faltas: {}".format(TotalFaltas))
```

```python
#Total de presenças

TotalPresencas = fimDeSemana["Alunos presentes"].sum() + diasUteis["Alunos presentes"].sum()

print("Total de presenças: {}".format(TotalPresencas))
```

```python
#Porcentagem de faltas

PorcentagemFaltas = (TotalFaltas / TotalPresencas) * 100

print("Porcentagem de faltas: {:.2f}%".format(PorcentagemFaltas))
```

```python
#Qual foi o dia da semana que teve mais faltas?

diaComMaisFaltas = planilha["Alunos com falta"].idxmax()

print("O dia da semana que teve mais faltas foi: {}".format(diaComMaisFaltas))
print ("Com um total de: {}".format(planilha["Alunos com falta"].max()))
```

```python
#Qual foi o dia da semana que teve mais presencas?

diaComMaisPresencas = planilha["Alunos presentes"].idxmax()

print("O dia da semana que teve mais presencas foi: {}".format(diaComMaisPresencas))
print ("Com um total de: {}".format(planilha["Alunos presentes"].max()))
```
